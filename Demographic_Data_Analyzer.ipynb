```
data = pd.read_csv('/content/drive/My Drive/adult_train.csv')
#Data can be accessed here: https://www.kaggle.com/datasets/sagnikpatra/uci-adult-census-data-dataset
data
```

```
race_count = data['Race'].value_counts()

average_age_men = data[data['Sex'] == 'Male']['Age'].mean()

percentage_bachelors = (data['Education'] == 'Bachelors').mean() * 100

advanced_education = ['Bachelors', 'Masters', 'Doctorate']
higher_education = data['Education'].isin(advanced_education)
percentage_high_earning = (data[higher_education]['Target'] == '>50K').mean() * 100

no_higher_education = ~higher_education
percentage_no_high_earning = (data[no_higher_education]['Target'] == '>50K').mean() * 100

min_work_hours = data['Hours_per_week'].min()

num_min_workers = data[data['Hours_per_week'] == min_work_hours]
rich_percentage = (num_min_workers['Target'] == '>50K').mean() * 100

country_stats = data['Country'].value_counts()
country_earnings = data[data['Target'] == '>50K']['Country'].value_counts()
highest_earning_country = (country_earnings / country_stats).idxmax()
highest_earning_country_percentage = (country_earnings / country_stats).max() * 100

indian_occupation_stats = data[(data['Country'] == 'India') & (data['Target'] == '>50K')]['Occupation'].value_counts()

print("Task 1: Race count:\n", race_count)
print("\nTask 2: Average age of men:", average_age_men)
print("\nTask 3: Percentage with a Bachelor's degree:", percentage_bachelors)
print("\nTask 4: Percentage with advanced education earning >50K:", percentage_high_earning)
print("\nTask 5: Percentage without advanced education earning >50K:", percentage_no_high_earning)
print("\nTask 6: Minimum work hours per week:", min_work_hours)
print("\nTask 7: Percentage of minimum work hours with >50K:", rich_percentage)
print("\nTask 8: Country with the highest percentage earning >50K:", highest_earning_country)
```
